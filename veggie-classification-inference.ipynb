# Model Prediction

```python
def prediction(file):
  img = tf.keras.utils.load_img(file, target_size=(255, 255))
  x = tf.keras.utils.img_to_array(img)

  plt.imshow(x.astype('uint8'))

  x = np.expand_dims(x, axis=0)

  images = np.vstack([x])
  classes = model.predict(images, batch_size=10)
  idx = np.argmax(classes)
  clas = ['Bean',
 'Bitter_Gourd',
 'Bottle_Gourd',
 'Brinjal',
 'Broccoli',
 'Cabbage',
 'Capsicum',
 'Carrot',
 'Cauliflower',
 'Cucumber',
 'Papaya',
 'Potato',
 'Pumpkin',
 'Radish',
 'Tomato']
  print('This is a {}'.format(clas[idx]))
```
